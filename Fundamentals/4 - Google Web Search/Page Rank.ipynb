```python
import networkx as nx
import matplotlib.pyplot as plt

def generate_graph(connections):
  G = nx.DiGraph()
  for n1, n2 in connections.items():
    G.add_nodes_from([n1, n2])
    G.add_edge(n1, n2)
  return G


def plot_graph(G, **kwargs):
  plt.figure(figsize=(16,8))
  plt.subplot(122)
  nx.draw_shell(G,**kwargs)
  plt.subplot(121)
  nx.draw(G, **kwargs)
```

```python
connections = {
  'A':'B',
  'A':'C',
  'A':'D',
  'C':'A',
  'C':'D',
  'D':'A',
  'D':'C',
  'B':'D',
}

G = generate_graph(connections)
plot_graph(G)
```
