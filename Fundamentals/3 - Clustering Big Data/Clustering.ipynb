```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import numpy as np
```

```python
test_cases = {}

for i in range(1,4):
  with open(f'Test Cases/testcase_{i}.txt', 'r') as file:
    file = file.read().split('\n')
    header = file.pop(0)
    test_cases[i] = {}
    test_cases[i].update({'Clusters': int(header[0])})
    
    points = []
    for line in file:
      p1, p2 = line.split('\t')
      points.append((float(p1), float(p2)))
    test_cases[i].update({'Points': points})
```

```python
def graph_nth_testcase(nth):
  clusters = test_cases[nth]['Clusters']
  X = test_cases[nth]['Points']
  kmeans = KMeans(n_clusters=clusters, random_state=0).fit(X)
  y_pred = kmeans.predict(X)

  x = [x for x,y in X]
  y = [y for x,y in X]
  plt.scatter(x, y, c=y_pred);
  
  centroids = kmeans.cluster_centers_
  points = list(kmeans.labels_)
  for i in range(clusters):
    print(f'Group {i} points: {points.count(i)}')
  print(f'\nCentroids:\n{centroids}')
```

```python
graph_nth_testcase(1)
```

```python
graph_nth_testcase(2)
```

```python
graph_nth_testcase(3)
```
